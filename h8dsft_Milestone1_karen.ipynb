- Measurable: Meningkatkan total penjualan sebesar 35% pada Q4 2018 dibandingkan dengan Q4 2017, menggunakan metrik seperti total revenue dan jumlah unit terjual.

- Achievable: Mengembangkan strategi pemasaran dan penjualan yang berbasis data, termasuk promosi produk yang diidentifikasi sebagai yang paling menguntungkan serta penyesuaian stok berdasarkan analisis penjualan sebelumnya.

- Time-bound: Mencapai peningkatan penjualan sebesar 35% pada akhir Q4 2018, dengan penyelesaian analisis dan rekomendasi strategi sebelum awal kuartal tersebut.

## Check Primary and Foreign Key

Pada tabel sales terdapat foreign key `product_id` dan `store_id` yang berhubungan dengan tabel products dan stores. Perlu dilakukan pengecekan untuk memastikan bahwa **semua product_id dan store_id di tabel sales memiliki pasangan yang valid di tabel terkait**. Dengan memastikan kesesuaian ini, dapat menjamin bahwa proses analisis yang membutuhkan tabel terkait akan aman, baik dalam hal retrieve data maupun dalam analisis lebih lanjut.

### Tabel Sales

```python
# Menghitung jumlah Product_ID pada sales yang tidak ada di products
count_false = (~sales['Product_ID'].isin(products['Product_ID'])).sum()

# Menampilkan jumlah Product_ID yang tidak ditemukan dalam tabel products
print(f'Jumlah Product_ID yang tidak ada di products: {count_false}')
```

Berdasarkan hasil pengecekan, tidak ada Product_ID yang bernilai False, yang berarti **semua Product_ID di tabel sales cocok dengan Product_ID di tabel product**. Hal ini memastikan bahwa hubungan antara primary key dan foreign key terjaga dengan baik, serta menunjukkan integritas data yang tinggi. Dengan demikian, data bisa untuk analisis.

```python
# Menghitung jumlah Store_ID pada sales yang tidak ada di stores
count_false = (~sales['Store_ID'].isin(stores['Store_ID'])).sum()

# Menampilkan jumlah Store_ID yang tidak ditemukan dalam tabel stores
print(f'Jumlah Store_ID yang tidak ada di stores: {count_false}')
```

## Set Price Values

### Set price value for product_price column

```python
# Mengambil nilai unik dari kolom 'product_price' dalam tabel 'products'
products['product_price'].unique()
```

Value price pada kolom product_price sebelumnya masih terdapat '$', pemisah ribuan (.), dan whitespace

```python
# Menghapus whitespace dan simbol '$' pada kolom 'product_price'
# Kemudian mengubah tipe data menjadi float dan membulatkannya hingga 2 desimal
products['product_price'] = products['product_price'].str.strip().replace({r'\$': ''}, regex=True).astype(float).round(2)
```

## [VIZ] Pada minggu ke berapa orang sering membeli barang?

```python
# Menambahkan kolom minggu dalam bulan
sales['week'] = (sales['date'].dt.day - 1) // 7 + 1  # Menentukan minggu 1-5 dalam bulan

# Menggabungkan tabel Sales dan Products untuk menghitung revenue
sales_product = pd.merge(sales, products[['product_id', 'product_price']], on='product_id')

# Menghitung revenue (total pendapatan)
sales_product['revenue'] = sales_product['units'] * sales_product['product_price']

# Menghitung total revenue per kuartal dan per minggu
total_revenue_by_week_quarter = sales_product.groupby(['quarter', 'week'])['revenue'].sum().reset_index()

# Menentukan posisi bar untuk setiap minggu
weeks = total_revenue_by_week_quarter['week'].unique()
bar_width = 0.15  # Lebar bar
indices = np.arange(len(total_revenue_by_week_quarter['quarter'].unique()))

# Membuat visualisasi total revenue per kuartal per minggu
plt.figure(figsize=(12, 6))

for i, week in enumerate(weeks):
    week_data = total_revenue_by_week_quarter[total_revenue_by_week_quarter['week'] == week]
    plt.bar(indices + i * bar_width, week_data['revenue'], width=bar_width, label=f'Minggu {week}')

plt.title('Total Revenue per Kuartal per Minggu')
plt.xlabel('Kuartal')
plt.ylabel('Total Revenue')
plt.xticks(indices + bar_width * (len(weeks) - 1) / 2, total_revenue_by_week_quarter['quarter'].unique())
plt.legend(title='Minggu')
plt.grid(axis='y')  # Menambahkan grid pada sumbu y
plt.tight_layout()  # Mengatur layout agar tidak terpotong
plt.show()
```

Secara keseluruhan, **minggu pertama**  dan **minggu ketiga** sering mencatat penjualan tinggi karena momen spesial dan perubahan musiman yang mendorong belanja. Namun, jika dilihat secara kasat mata, distribusi minggu ke 1 hingga 4 (diasumsikan minggu 5 hanya beberapa hari), memiliki penyebaran yang sama. Oleh karena itu, untuk membuktikan apakah penjualan untuk setiap minggu berpengaruh dengan revenue, dilakukan pengujian tambahan menggunakan ANOVA. ANOVA digunakan untuk menguji setiap minggu pada setiap kuartal apakah berpengaruh dengan revenue.

```python
from scipy import stats

# Menggabungkan tabel Sales dan Products untuk menghitung revenue
sales_week = pd.merge(sales, products[['product_id', 'product_price']], on='product_id')

# Menghitung revenue (total pendapatan)
sales_week['revenue'] = sales_week['units'] * sales_week['product_price']

# Lakukan gruping berdasarkan minggu untuk mendapatkan total revenue per minggu
revenue_per_week = sales_week.groupby('week')['revenue'].sum().reset_index()

# Kelompokkan data total revenue per minggu per kuartal
weekly_revenue = sales_week.groupby(['quarter', 'week'])['revenue'].sum().reset_index()

# Lakukan ANOVA berdasarkan minggu per kuartal
anova_result = stats.f_oneway(
    *[weekly_revenue[weekly_revenue['quarter'] == quarter]['revenue']
      for quarter in weekly_revenue['quarter'].unique()]
)

# Hasil ANOVA
print(f'F-statistic: {anova_result.statistic}')
print(f'P-value: {anova_result.pvalue}')

# Interpretasi hasil
if anova_result.pvalue < 0.05:
    print("Terdapat perbedaan signifikan antara revenue di minggu yang berbeda antar kuartal.")
else:
    print("Tidak terdapat perbedaan signifikan antara revenue di minggu yang berbeda antar kuartal.")
```

Hasil ANOVA menunjukkan nilai **F-statistic** sebesar **1.156** dan **P-value** sebesar **0.357**. Karena **nilai p** lebih besar dari 0.05, dapat disimpulkan bahwa **tidak terdapat perbedaan signifikan** antara revenue di minggu yang berbeda antar kuartal. Ini berarti bahwa berdasarkan data yang dianalisis, pola revenue mingguan tidak berbeda secara signifikan antara kuartal yang dibandingkan.

## [DESC] Apa mean, median, standar deviasi, skewness, dan kurtosis dari harga produk, units, dan revenue?

```python
# Gabungkan sales dan product
merged_df = pd.merge(sales, products, on='product_id')

merged_df
# Hitung revenue
merged_df['revenue'] = merged_df['product_price'] * merged_df['units']

# Hitung statistik deskriptif
statistics = {
    'product_price': {
        'mean': merged_df['product_price'].mean(),
        'median': merged_df['product_price'].median(),
        'std_dev': merged_df['product_price'].std(),
        'skewness': merged_df['product_price'].skew(),
        'kurtosis': merged_df['product_price'].kurtosis()
    },
    'units': {
        'mean': merged_df['units'].mean(),
        'median': merged_df['units'].median(),
        'std_dev': merged_df['units'].std(),
        'skewness': merged_df['units'].skew(),
        'kurtosis': merged_df['units'].kurtosis()
    },
    'revenue': {
        'mean': merged_df['revenue'].mean(),
        'median': merged_df['revenue'].median(),
        'std_dev': merged_df['revenue'].std(),
        'skewness': merged_df['revenue'].skew(),
        'kurtosis': merged_df['revenue'].kurtosis()
    }
}

# Menampilkan hasil
statistics_df = pd.DataFrame(statistics).T
statistics_df
```

- Nilai p yang sangat kecil (0.0) menunjukkan bahwa hubungan ini signifikan secara statistik. Ini berarti bahwa kemungkinan hubungan antara harga produk dan jumlah unit terjual terjadi hanya karena kebetulan sangat kecil.

## [INF] Apakah terdapat hubungan antara kategori produk dan kecenderungan pembelian dalam jumlah besar?

Untuk mengetahui apakah terdapat hubungan antara kategori produk dan kecenderungan pembelian dalam jumlah besar, perlu dilakukan uji hipotesis. Dalam konteks ini, akan menggunakan uji Chi-Square.

Uji Chi-Square dipilih karena dapat digunakan untuk menguji independensi antara dua variabel kategorikal. Dalam hal ini, kategori produk dan jumlah unit yang dibeli dapat dikategorikan.

Hipotesis:
- Hipotesis Nol (H0): Tidak ada hubungan antara kategori produk dan kecenderungan pembelian dalam jumlah besar.
  
- Hipotesis Alternatif (H1): Ada hubungan antara kategori produk dan kecenderungan pembelian dalam jumlah besar.

```python
from scipy.stats import chi2_contingency

# Menghitung median unit terjual
median_units = merged_df['units'].median()

# Membuat kolom baru untuk kategori pembelian
merged_df['purchase_category'] = ['Besar' if x > median_units else 'Kecil' for x in merged_df['units']]

# Membuat tabel kontingensi
contingency_table = pd.crosstab(merged_df['product_category'], merged_df['purchase_category'])

# Melakukan uji Chi-Square
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Menampilkan hasil
print("Chi-Square Statistic:", chi2)
print("Nilai p:", p)
if p < 0.05:
    print("Ada hubungan signifikan antara kategori produk dan kecenderungan pembelian dalam jumlah besar.")
else:
    print("Tidak ada hubungan signifikan antara kategori produk dan kecenderungan pembelian dalam jumlah besar.")
```

Ada hubungan signifikan antara kategori produk dan kecenderungan pembelian dalam jumlah besar. Ini menunjukkan bahwa kategori produk mempengaruhi bagaimana banyak unit yang dibeli oleh konsumen.

## [INF] Berapa estimasi minimal dan maksimal untuk kenaikan revenue berdasarkan kenaikan persentase antara Q3 2017 dan Q4 2017?

```python
# Mengambil revenue Q3 2017 dan Q4 2017
revenue_q3_2017 = quarterly_revenue.loc[quarterly_revenue['quarter'] == '2017Q3', 'revenue'].values[0]
revenue_q4_2017 = quarterly_revenue.loc[quarterly_revenue['quarter'] == '2017Q4', 'revenue'].values[0]

# Menghitung kenaikan persentase
percentage_increase = ((revenue_q4_2017 - revenue_q3_2017) / revenue_q3_2017) * 100

# Mengambil revenue Q3 2018
revenue_q3_2018 = quarterly_revenue.loc[quarterly_revenue['quarter'] == '2018Q3', 'revenue'].values[0]

# Menghitung estimasi revenue Q4 2018
estimated_revenue_min = revenue_q3_2018 * (1 + (percentage_increase * 0.9 / 100))  # Estimasi minimal
estimated_revenue_max = revenue_q3_2018 * (1 + (percentage_increase * 1.1 / 100))  # Estimasi maksimal

# Menghitung persentase kenaikan dari Q3 2018 ke estimasi minimal Q4 2018
percentage_increase_from_q3_to_estimated_min = ((estimated_revenue_min - revenue_q3_2018) / revenue_q3_2018) * 100
percentage_increase_from_q3_to_estimated_max = ((estimated_revenue_max - revenue_q3_2018) / revenue_q3_2018) * 100

# Menampilkan hasil
print('Estimasi Minimal Revenue Q4 2018:', estimated_revenue_min)
print('Estimasi Maksimal Revenue Q4 2018:', estimated_revenue_max)
print('Persentase Kenaikan dari Q3 2018 ke Estimasi Minimal Q4 2018:', percentage_increase_from_q3_to_estimated_min)
print('Persentase Kenaikan dari Q3 2018 ke Estimasi Maksimal Q4 2018:', percentage_increase_from_q3_to_estimated_max)
```

Berdasarkan estimasi, **revenue minimal untuk Q4 2018** diprediksi sekitar **Rp2,776,617.11** dengan persentase kenaikan sebesar **29.30%** dari Q3 2018. Sementara itu, **revenue maksimal** diperkirakan mencapai **Rp2,916,438.60**, dengan kenaikan **35.81%**.

Dengan demikian, target kenaikan revenue sebesar 35% sangat memungkinkan karena masih berada dalam rentang estimasi ini.
